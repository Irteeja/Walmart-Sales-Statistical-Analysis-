## Walmart Sales Statistical Analysis
### We willll calculate summary statistics on DataFrame columns, and use grouped summary statistics and pivot tables.

```python
import pandas as pd
import numpy as np
```

```python
walmart_sales=pd.read_csv("sales_subset.csv")
walmart_sales.head()
```

```python
#Print information about the columns in sales
walmart_sales.info()
#Print the mean of the weekly_sales column.
walmart_sales["weekly_sales"].mean()
#Print the median of the weekly_sales column.
walmart_sales["weekly_sales"].median()
```

```python
# Print the maximum of the date column.
print(walmart_sales["date"].max())
# Print the minimum of the date column.
print(walmart_sales["date"].min())
```

```python
#Use the custom iqr function defined for you along with .agg() to print the IQR of the temperature_c column of sales.
def iqr(column):
    return column.quantile(0.75)-column.quantile(0.25)
walmart_sales["temperature_c"].agg(iqr)
```

```python
sales_1_1=walmart_sales[(walmart_sales["store"]==1) & (walmart_sales["department"]==1)]
sales_1_1.head()
```

```python
#Sort the rows of sales_1_1 by the date column in ascending order.
sales_1_1=sales_1_1.sort_values("date",ascending=True)
sales_1_1
```

```python
#Get the cumulative sum of weekly_sales and add it as a new column of sales_1_1 called cum_weekly_sales.
sales_1_1["cum_weekly_sales"]=sales_1_1["weekly_sales"].cumsum()
sales_1_1.head()
```

```python
#Get the cumulative maximum of weekly_sales, and add it as a column called cum_max_sales.
sales_1_1["cum_max_sales"]=sales_1_1["weekly_sales"].cummax()
sales_1_1.head()
```

```python
#Print the date, weekly_sales, cum_weekly_sales, and cum_max_sales columns.
sales_1_1[["date","weekly_sales","cum_weekly_sales","cum_max_sales"]]
```

### Dropping Duplicates

```python
#Remove rows of sales with duplicate pairs of store and type and save as store_types and print the head.
store_types=walmart_sales.drop_duplicates(subset=["store","type"])
store_types
```

```python
#Remove rows of sales with duplicate pairs of store and department and save as store_depts and print the head.
store_depts=walmart_sales.drop_duplicates(subset=["store","department"])
store_depts
```

```python
#Subset the rows that are holiday weeks using the is_holiday column, and drop the duplicate dates, saving as holiday_dates.
holiday_dates=walmart_sales[walmart_sales["is_holiday"]==True].drop_duplicates("date")
holiday_dates
```

```python
#Select the date column of holiday_dates, and print.
holiday_dates["date"]
```

### Counting Categorical variables

```python
#Count the number of stores of each store type in store_types.
store_counts=store_types["type"].value_counts()
store_counts
```

```python
#Count the proportion of stores of each store type in store_types.
store_props=store_types["type"].value_counts(normalize=True)
store_props
```

```python
#Count the number of different departments in store_depts, sorting the counts in descending order
depts_counts_sorted=store_depts["department"].value_counts(sort=True)
depts_counts_sorted
```

```python
# Get the proportion of departments of each number and sort
dept_props_sorted = store_depts["department"].value_counts(sort=True, normalize=True)
print(dept_props_sorted)
```

```python
# Calc total weekly sales
sales_all = walmart_sales["weekly_sales"].sum()

# Subset for type A stores, calc total weekly sales
sales_A = walmart_sales[walmart_sales["type"] == "A"]["weekly_sales"].sum()

# Subset for type B stores, calc total weekly sales
sales_B = walmart_sales[walmart_sales["type"]=="B"]["weekly_sales"].sum()

# Subset for type C stores, calc total weekly sales
sales_C = walmart_sales[walmart_sales["type"]=="C"]["weekly_sales"].sum()

# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
# Group by type; calc total weekly sales
sales_by_type = walmart_sales.groupby("type")["weekly_sales"].sum()
```

```python
#Calculate the proportion of sales at each store type by dividing by the sum of sales_by_type. Assign to sales_propn_by_type.
sales_propn_by_type=sales_by_type/sum(walmart_sales.weekly_sales)
sales_propn_by_type
```

```python
sales_by_type = walmart_sales.groupby("type")["weekly_sales"].sum()
sales_by_type_is_holiday = walmart_sales.groupby(["type", "is_holiday"])["weekly_sales"].sum()
```

```python
sales_by_type
```

```python
sales_by_type_is_holiday
```

```python
#Get the min, max, mean, and median of weekly_sales for each store type using .groupby() and .agg(). 
#Store this as sales_stats. Make sure to use numpy functions!
sales_stats=walmart_sales.groupby("type")["weekly_sales"].agg([np.max,np.min,np.mean,np.median])
```

```python
sales_stats
```

```python
#Get the min, max, mean, and median of unemployment and fuel_price_usd_per_l for each store type. Store this as unemp_fuel_stats.
unemp_fuel_stats=walmart_sales.groupby("type")[["unemployment","fuel_price_usd_per_l"]].agg([np.max,np.min,np.median,np.mean])
```

```python
unemp_fuel_stats
```

### Pivot Tables

```python
#Get the mean weekly_sales by type using .pivot_table() and store as mean_sales_by_type.
walmart_sales.pivot_table(values="weekly_sales", index="type")
```

```python
#Get the mean and median (using NumPy functions) of weekly_sales by type using .
#pivot_table() and store as mean_med_sales_by_type.
mean_med_sales_by_type=walmart_sales.pivot_table(values="weekly_sales",index="type",aggfunc=([np.mean,np.median]))
```

```python
mean_med_sales_by_type
```

```python
#Get the mean of weekly_sales by type and is_holiday using .pivot_table() and store as mean_sales_by_type_holiday.
mean_sales_by_type_holiday=walmart_sales.pivot_table(columns="is_holiday",values="weekly_sales",index="type")
```

```python
mean_sales_by_type_holiday
```

### Fill in missing values and sum values with pivot tables

```python
#Print the mean weekly_sales by department and type, filling in any missing values with 0
print(walmart_sales.pivot_table(columns="type",index="department",values="weekly_sales",fill_value=0))
```

```python
# Print the mean weekly_sales by department and type; fill missing values with 0s; sum all rows and cols
print(walmart_sales.pivot_table(values="weekly_sales", index="department", columns="type", fill_value=0,margins=True))
```
